```python
P = 1440 # one day in minuts

DAYS_MONTHS = 30
total_months = 4
training_months = 2
test_months = total_months - training_months

training = training_months*DAYS_MONTHS 
test = test_months*DAYS_MONTHS
training = 

M = np.array(P*[total_months*DAYS_MONTHS]) # total nb of servers can be on in a minut, 3 months times 30 days 
Y = np.array(server_2['on/off'].tolist()).reshape(total_months*DAYS_MONTHS,P).sum(axis=0)
Y_test = np.array(server_2['on/off'].tolist()).reshape(total_months*DAYS_MONTHS,P)[training:,:].sum(axis=0)
T = np.array(list(range(1,1441)))

print(training,test)
print(Y.shape)
print(Y_test.shape)
print(T)
```

```python
C = lambda t: math.cos(t*2*math.pi/P)
S = lambda t: math.sin(t*2*math.pi/P)

X =  np.array([[1,C(t),S(t)] for t in T])

#X =  np.array([[C(t),S(t)] for t in T])


RES = Y/(training+test)
print(RES)
log_reg = sm.Logit(RES, X).fit(method="newton")
print("params",log_reg.params)
print(log_reg.summary())
```

```python
# performing predictions on the test dataset
T = range(300,320)
yhat = [log_reg.predict([1,C(t),S(t)]) for t in T]
prediction = list([x[0] for x in yhat])
  
# comparing original and predicted values of y
print('Actual values', list(Y_test/test)[300:320])
print('Predictions :', prediction)
```
