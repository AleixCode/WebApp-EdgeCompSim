```python
import numpy as np
import math
import statsmodels.api as sm
```

```python
T=np.array([6.,8.,10.,12.5,14.0,15.5,18.,20.,22.])
M=np.array([36,80,81,108,60,35,69,68,68])
Y=np.array([10,6,7,2,2,6,14,14,36])
m=0
x=np.empty([605,3])
y=np.empty(605)
```

```python
print(x)
print(y)
```

```python
x.shape
```

```python
for i in range(9):
    if i==0:
        k=0
    else:
        k=k+M[i-1]
    for j in range(M[i]):
        x[j+k,:]=[1., math.cos(T[i]*2.*math.pi/24.), math.sin(T[i]*2.*math.pi/24.)]
#         print(j+k,x[j+k,:])
```

```python
x
```

```python
for i in range(9):
    count=0
    if i==0:
        k=0
    else:
        k=k+M[i-1]
    for j in range(M[i]):
        x[j+k,:]=[1., math.cos(T[i]*2.*math.pi/24.), math.sin(T[i]*2.*math.pi/24.)]
        if count<Y[i]:
            y[j+k]=1
            count=count+1
            print( j+k, x[j+k,:], count, y[j+k])
        else:
            y[j+k]=0
            print( j+k, x[j+k,:], count, y[j+k])
```
